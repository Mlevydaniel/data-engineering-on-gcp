```python
%sql

CREATE TEMPORARY VIEW orders (
    order_id INT,
    order_date STRING,
    order_customer_id INT,
    order_status STRING
) USING CSV
OPTIONS (
    path='dbfs:/public/retail_db/orders/'
)
```

```python
%sql

SELECT * FROM orders
```

```python
%sql

SELECT order_status, count(*) AS order_count 
FROM orders
GROUP BY 1
ORDER BY 2 DESC
```
